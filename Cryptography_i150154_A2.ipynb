```python
'''shift(Ar) basically is a shift right function that takes a list or string and perform shift right once
then return an updated list'''
def shift(Ar):
    b=[0 for i in range(len(Ar))]
    b[0]=Ar[len(Ar)-1]
    for i in range(len(Ar)-1):
        b[i+1]=Ar[i]
    return b
```

```python
'''andof2(Ar,ind1,ind2) is a function that takes two indexes with a list or string and performs and(&)
 between those values and then returns resultant value'''
def andof2(Ar,ind1,ind2):
    test=0
    if Ar[ind1-1]+Ar[ind2-1]==2:
        test=1
    else:
        test=0
    return test
```

```python
'''XORof2(a,b) is a function that takes two integer values and performs XOR
 between those values and then returns resultant value'''
def XORof2(a,b):
    test=0
    if a==b:
        test=0
    else:
        test=1
    return test
```

```python
'''XORof3(a,b,c) is a function that takes three integer values and performs XOR
 between those values and then returns resultant value'''
def XORof3(a,b,c):
    test=0
    if a+b+c==1 or a+b+c==3:
        test=1
    else:
        test=0
    return test
```

```python
'''trivium(PT) generates a key with 0 as clock and gives a key of length of Plain Text
to be Encrypted or Decrypted'''
def trivium(PT):
    '''Initializing three registers and creating list for KeyStream'''
    A = [0 for i in range(93)]
    B = [0 for i in range(84)]
    C = [0 for i in range(111)]
    C[108] = C[109] = C[110] = 1
    S = [0 for i in range(len(PT))]
    '''X1->6 are variables to store values returned by respective XOR fuction while and1->3 stores anded values'''
    X1=0
    X2=0
    X3=0
    X4=0
    X5=0
    X6=0
    and1=0
    and2=0
    and3=0
    '''Loop to generate Key of the same length as of PlainText through procedure defined by Trivium Cipher'''
    for i in range(len(PT)):
        and1=andof2(A,91,92)
        and2=andof2(B,82,83)
        and3=andof2(C,109,110)
        X1=XORof3(A[65],and1,A[92])
        X2=XORof3(B[68],and2,B[83])
        X3=XORof3(C[65],and3,C[110])
        X4=XORof2(X3,A[68])
        X5=XORof2(X1,B[77])
        X6=XORof2(X2,C[86])
        A=shift(A)
        B=shift(B)
        C=shift(C)
        A[0]=X4
        B[0]=X5
        C[0]=X6
        S[i]=XORof3(X1,X2,X3)
    return S
```

```python
'''encrypt(PT,KS) encrypts PlainText(PT) with KeyStream(KS) by XORing bit by bit operation 
to generate CipherText(CT)'''
def encrypt(PT,KS):
    CT = [0 for i in range(len(PT))]
    for i in range(len(PT)):
        CT[i]=XORof2(int(PT[i]),int(KS[i]))
    return CT
```

```python
'''decrypt(CT,KS) encrypts CipherText(CT) with KeyStream(KS) by XORing bit by bit operation 
to generate PlainText(PT)'''
def decrypt(CT,KS):
    PT = [0 for i in range(len(CT))]
    for i in range(len(CT)):
        PT[i]=XORof2(int(CT[i]),int(KS[i]))
    return PT
```

```python
'''converter(st) is used in encryption and converts given string to its ASCII binary values and scale them 
to 7 bits by calling scaledata(st) and then returns the list with bits converted from input characters'''
def converter(st):
    li=[]
    for i in range(len(st)):
        x=scaledata(st[i])          #sends 1 character at a time to get scaled and gets 7 bit binary value
        for i in range(len(x)):
            li.append(x[i])
    return li
```

```python
'''scaledata(st) scales a string type character into a 7 bit binary ASCII and returns in the form of list
it is used in encryption process'''
def scaledata(st):
    x=''.join(format(ord(x), 'b') for x in st)     #converts given string to binary
    temp=[0,0,0,0,0,0,0]
    if len(x)<7:                                   #in case function provides less than 7 bits e.g. for "0"
        j=-1
        for i in range(len(x)):
            temp[j]=x[j]
            j-=1
        return temp
    else:
        li=[]
        for i in range(len(x)):
            li.append(x[i])
        return li

```

```python
'''convert(st) converts string(st) containing Ciphered Value for decryption into ASCII Decimal values'''
def convert(st):
    ind=0
    to=[]
    for i in range(len(st)/7):                    #as each 7 bits represents one character
        li=[]
        for j in range(7):
            li.append(st[ind+j])
        ind=ind+7
        x=scale(li)                               #returns Decimal ASCII value
        to.append(x)
    return to
```

```python
'''scale(st) scales given 7 bits to Decimal value ranging from 0-127 for decryption'''
def scale(st):
    p = ""
    for i in st:
        p+=str(i)
    return int(p,2)
```

```python
'''decipher(st) takes list as an input an returns corresponding character thrugh ASCII Table'''
def decipher(st):
    f=[]
    for i in range(len(st)):
        f.append(chr(st[i]))
    return f
```

```python
'''Generate(bits,clk) provides the KeyStream via Trivium implementation with size of (bits) and
warm up clocked (clk) times'''
def Generate(bits,clk):
    '''Initializing three registers and creating list for KeyStream'''
    A = [0 for i in range(93)]
    B = [0 for i in range(84)]
    C = [0 for i in range(111)]
    C[108] = C[109] = C[110] = 1
    S = [0 for i in range(bits)]
    '''X1->6 are variables to store values returned by respective XOR fuction while and1->3 stores anded values'''
    X1=0
    X2=0
    X3=0
    X4=0
    X5=0
    X6=0
    and1=0
    and2=0
    and3=0
    
    indcheck=0                                 #Enables S to start storing Key bits
    '''Loop to generate Key of size (bits) and clocked (clk) times through procedure defined by Trivium Cipher'''
    for i in range(bits+clk):
        and1=andof2(A,91,92)
        and2=andof2(B,82,83)
        and3=andof2(C,109,110)
        X1=XORof3(A[65],and1,A[92])
        X2=XORof3(B[68],and2,B[83])
        X3=XORof3(C[65],and3,C[110])
        X4=XORof2(X3,A[68])
        X5=XORof2(X1,B[77])
        X6=XORof2(X2,C[86])
        A=shift(A)
        B=shift(B)
        C=shift(C)
        A[0]=X4
        B[0]=X5
        C[0]=X6
        if i>=clk:
            S[indcheck]=XORof3(X1,X2,X3)
            indcheck=indcheck+1
    return S
```
